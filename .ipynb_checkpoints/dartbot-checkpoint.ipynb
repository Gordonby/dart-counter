```python
import datetime

def get_iso_day_range(date_str):
    """
    Given a date string (YYYY-MM-DD),
    return a tuple of two ISO timestamps:
      1) Start of the given day (00:00:00.000Z)
      2) Start of the next day (00:00:00.000Z),
         effectively covering midnight-to-midnight.
    """
    # Parse the input date (no time info, defaults to 00:00:00)
    date_obj = datetime.datetime.strptime(date_str, "%Y-%m-%d")

    # Start of the day (already 00:00:00)
    start_of_day = date_obj
    
    # Start of the next day
    end_of_day = start_of_day + datetime.timedelta(days=1)

    # Format them: YYYY-MM-DDTHH:MM:SS.mmmZ
    # We use strftime with microseconds, then slice to 3 decimals, then append 'Z'.
    def to_iso(dt):
        iso_str = dt.strftime("%Y-%m-%dT%H:%M:%S.%f")  # e.g. 2024-01-09T00:00:00.000000
        return iso_str[:-3] + "Z"                      # => 2024-01-09T00:00:00.000Z

    return to_iso(start_of_day), to_iso(end_of_day)

start, end = get_iso_day_range("2025-01-09")
print("Start of day:", start)  # e.g. 2025-01-09T00:00:00.000Z
print("End of day:  ", end)    # e.g. 2025-01-10T00:00:00.000Z
```

```python
import requests
import datetime
import json

headers = {
    "Authorization": f"Bearer {access_token}",
    "accept": "application/json",
    "content-type": "application/json",
}


def fetch_matches_for_date(input_date):
    # input date is in format YYYY-MM-DD
    from_date, to_date = get_iso_day_range(input_date)
    
    # 4. Set up the query parameters
    params = {
        "user_id": user["id"],
        "from_date": from_date,
        "to_date": to_date,
        "is_verified": True,
        "is_online": False,
        "limit": 5,
        "page": 1
        # You can also pass limit, page, is_verified, is_online, etc. if desired
    }
    
    # 5. Define the endpoint
    url = "https://api.dartcounter.net/matches"
    
    all_matches = []
    page = 1
    page_size = 5
    last_page = 1
    
    while page <= last_page:
        print(f"Requesting page {page}, with {page_size} per page")
        params = {
            "from_date": from_date,
            "to_date": to_date,
            "page": page,
            "limit": 10,   # Or any number the API allows
        }
        
        response = requests.get("https://api.dartcounter.net/matches", headers=headers, params=params)
    
        if response.status_code != 200:
            print(f"Error {response.status_code}: {response.text}")
            break
    
        data = response.json()
        
        # The "data" key often holds the array of match objects
        # but the pagination details are at the top level of the JSON.
        matches_on_page = data.get("data", [])
        for match in matches_on_page:
            for match_user in match["users"]:
                del match_user["user"]["profile"]["birthdate"]
        all_matches.extend(matches_on_page)
        
        current_page = data.get("current_page", 1)
        last_page = data.get("last_page", 1)
        
        print(f"Fetched {len(matches_on_page)} matches on page {current_page} of {last_page}.")
        
        page += 1
    
    print(f"Fetched {len(all_matches)} matches")
    return all_matches
```

```python
now = datetime.datetime.now()
yesterday_dt = now - datetime.timedelta(days=1)
yesterday = yesterday_dt.strftime("%Y-%m-%d")

yesterday_matches = fetch_matches_for_date(yesterday)
print(json.dumps(all_matches[0], indent=2))
```

```python
def is_match_valid(match_data):
    if not match_data["users"] or len(match_data["users"]) > 1:
        print(f"Match {match_id} has more than one user, skipping")
        return False
    if match_data["end_with"] != "double_out":
        print(f"Match didn't end in double out, skipping")
        return False
    if match_data["start_score"] != 501:
        print(f"Match didn't have 501 score, skipping")
        return False
    return True


def fetch_all_legs_in_matches(matches):
    all_legs = []
    
    for match in matches:
        match_id = match["id"]
        url = f"https://api.dartcounter.net/matches/{match_id}"
        
        response = requests.get(url, headers=headers)
        if response.status_code == 200:
            match_data = response.json()
            match_is_valid = is_match_valid(match_data)
            if not match_is_valid:
                continue
            
            all_legs.extend(match_data["legs"])
            num_legs = len(match_data["legs"])
            print(f"Match {match_id} had {num_legs} legs")
            # Optionally do additional processing here
        else:
            print(f"Error fetching match {match_id}: {response.status_code} - {response.text}")
    
    print(f"Fetched details for {len(all_legs)} legs")
    return all_legs
```

```python
yesterday_legs = fetch_all_legs_in_matches(yesterday_matches)
print(json.dumps(yesterday_legs[1], indent=2))
```

```python
def analyse_leg(leg):
    """
    Analyze a single-leg dictionary under these assumptions:
      1) One user per leg
      2) Leg is always completed (remaining_score == 0 on last turn)
    Returns a dict of computed stats.
    """

    # Extract all the turns
    turns = leg["turns"]  # Guaranteed to exist and be non-empty

    # ---------------------
    # 1) Three-Dart Average
    # ---------------------
    total_scored = 0
    total_darts_thrown = 0

    for turn in turns:
        total_scored += turn["total_score"]
        total_darts_thrown += turn["darts_thrown"]  # might be 13

    # Standard 3-dart average
    three_dart_avg = 0
    if total_darts_thrown > 0:
        three_dart_avg = (total_scored / total_darts_thrown) * 3

    # -----------------
    # 2) First 9 Darts
    # -----------------
    # "First 9" means first 3 turns, if each is up to 3 darts.
    first_3_turns = turns[:3]
    first_9_scored = sum(t["total_score"] for t in first_3_turns)
    total_first_9_darts = sum(t["darts_thrown"] for t in first_3_turns)

    first_9_avg = 0
    if total_first_9_darts > 0:
        first_9_avg = (first_9_scored / total_first_9_darts) * 3

    # ----------------
    # 3) Highest Finish
    # ----------------
    # With one user & guaranteed finish, the last turn is the finishing turn.
    # That finishing "checkout" is the last turn's total_score.
    highest_finish = turns[-1]["total_score"]

    # --------------------------------
    # 4) Checkout Opportunities (throws)
    # --------------------------------
    # If "is_opportunity": 1 means a checkout opportunity.
    checkout_opportunities = 0
    for turn in turns:
        for dart_throw in turn["throws"]:
            if dart_throw.get("is_opportunity") == 1:
                checkout_opportunities += 1

    # --------------------------------------------
    # 5) High Scores per Turn (180, 160+, 140+ etc)
    # --------------------------------------------
    count_180 = 0
    count_160_plus = 0
    count_140_plus = 0
    count_120_plus = 0
    count_100_plus = 0
    count_80_plus = 0
    count_60_plus = 0
    count_40_plus = 0
    count_under_40 = 0

    for turn in turns:
        score = turn["total_score"]
        if score == 180:
            count_180 += 1
            continue
        if score >= 160:
            count_160_plus += 1
            continue
        if score >= 140:
            count_140_plus += 1
            continue
        if score >= 120:
            count_120_plus += 1
            continue
        if score >= 100:
            count_100_plus += 1
            continue
        if score >= 80:
            count_80_plus += 1
            continue
        if score >= 60:
            count_60_plus += 1
            continue
        if score >= 40:
            count_40_plus += 1
            continue
        count_under_40 += 1

    return {
        "three_dart_avg": three_dart_avg,
        "first_9_avg": first_9_avg,
        "highest_finish": highest_finish,
        "checkout_opportunities": checkout_opportunities,
        "count_180": count_180,
        "count_160_plus": count_160_plus,
        "count_140_plus": count_140_plus,
        "count_120_plus": count_120_plus,
        "count_100_plus": count_100_plus,
        "count_80_plus": count_80_plus,
        "count_60_plus": count_60_plus,
        "count_40_plus": count_40_plus,
        "count_under_40": count_under_40
    }
```

```python
most_recent_leg_stats = analyse_leg(yesterday_legs[-1])
print(json.dumps(most_recent_leg_stats, indent=2))
```

```python
def aggregate_legs(legs):
    """
    Aggregate stats from a list of legs. 
    Each leg is expected to have:
        - "turns": list of turn dicts
          - turn["total_score"]
          - turn["darts_thrown"]
          - turn["remaining_score"] (to detect finishing turn)
          - turn["throws"] (list of throws, each with possible is_opportunity==1)
    
    Returns a dictionary with:
        - global_3_dart_avg
        - global_first_9_avg
        - highest_finish_all
        - checkout_percentage
        - total_checkout_opportunities
        - total counts for 180, 160+, 140+, 120+, 100+, 80+, 60+, 40+, and under_40
    """

    # Summing raw data for a "true" overall average
    total_scored_all = 0
    total_darts_all = 0
    total_checkout_ops_all = 0
    count_180_all = 0
    count_160_plus_all = 0
    count_140_plus_all = 0
    count_120_plus_all = 0
    count_100_plus_all = 0
    count_80_plus_all = 0
    count_60_plus_all = 0
    count_40_plus_all = 0
    count_under_40 = 0
    highest_finish_all = 0

    # For global first 9
    first_9_scored_all = 0
    first_9_darts_all = 0

    for leg in legs:
        turns = leg.get("turns", [])

        # Accumulate data for global 3-dart average
        for turn in turns:
            total_scored_all += turn["total_score"]
            total_darts_all += turn["darts_thrown"]

            # Check if this turn finished the leg (remaining_score=0) for highest finish
            if turn["remaining_score"] == 0:
                finish_value = turn["total_score"]
                if finish_value > highest_finish_all:
                    highest_finish_all = finish_value

            # Checkout opportunities
            for dart_throw in turn["throws"]:
                if dart_throw.get("is_opportunity") == 1:
                    total_checkout_ops_all += 1

            # Count high scores per turn
            score = turn["total_score"]
            if score == 180:
                count_180_all += 1
                continue
            if score >= 160:
                count_160_plus_all += 1
                continue
            if score >= 140:
                count_140_plus_all += 1
                continue
            if score >= 120:
                count_120_plus_all += 1
                continue
            if score >= 100:
                count_100_plus_all += 1
                continue
            if score >= 80:
                count_80_plus_all += 1
                continue
            if score >= 60:
                count_60_plus_all += 1
                continue
            if score >= 40:
                count_40_plus_all += 1
                continue
            count_under_40 += 1

        # Accumulate data for global first-9 (first 3 turns of each leg)
        first_3_turns = turns[:3]
        for turn in first_3_turns:
            first_9_scored_all += turn["total_score"]
            first_9_darts_all += turn["darts_thrown"]

    # Compute final aggregates
    if total_darts_all > 0:
        global_3_dart_avg = (total_scored_all / total_darts_all) * 3
    else:
        global_3_dart_avg = 0

    if first_9_darts_all > 0:
        global_first_9_avg = (first_9_scored_all / first_9_darts_all) * 3
    else:
        global_first_9_avg = 0

    if total_checkout_ops_all > 0:
        checkout_percentage = (len(legs) / total_checkout_ops_all) * 100
    else:
        checkout_percentage = 0

    return {
        "total_legs": len(legs),
        "total_darts": total_darts_all,
        "3_dart_avg": round(global_3_dart_avg, 2),
        "first_9_avg": round(global_first_9_avg, 2),
        "highest_finish": highest_finish_all,
        "checkout_percentage": round(checkout_percentage, 1),
        "total_checkout_opportunities": total_checkout_ops_all,
        "total_180": count_180_all,
        "180_per_leg": count_180_all/len(legs),
        "total_160_plus": count_160_plus_all,
        "160_per_leg": count_160_plus_all/len(legs),
        "total_140_plus": count_140_plus_all,
        "140_per_leg": count_140_plus_all/len(legs),
        "total_120_plus": count_120_plus_all,
        "120_per_leg": count_120_plus_all/len(legs),
        "total_100_plus": count_100_plus_all,
        "100_per_leg": count_100_plus_all/len(legs),
        "total_80_plus": count_80_plus_all,
        "80_per_leg": count_80_plus_all/len(legs),
        "total_60_plus": count_60_plus_all,
        "60_per_leg": count_60_plus_all/len(legs),
        "total_40_plus": count_40_plus_all,
        "40_per_leg": count_40_plus_all/len(legs),
        "total_under_40": count_under_40,
        "under_40_per_leg": count_under_40/len(legs),
    }
```

```python
yesterday_analysis = aggregate_legs(yesterday_legs)
print(json.dumps(yesterday_analysis, indent=2))
```

```python
def get_stats_for_day(date_str):
    m = fetch_matches_for_date(date_str)
    if len(m) == 0:
        return None
    l = fetch_all_legs_in_matches(m)
    if len(l) == 0:
        return None
    s = aggregate_legs(l)
    return s
```

```python
def get_last_n_days_stats(n=10):
    """
    Fetch stats for the last 'n' days, returning a list of dicts.
    Each dict contains:
      {
        "date": "YYYY-MM-DD",
        "stats": <your existing function's result>,
      }
    """
    results = []
    today = datetime.datetime.now()

    for i in range(1, n + 1):
        # e.g., 1 day ago, 2 days ago, ..., n days ago
        date_dt = today - datetime.timedelta(days=i)
        date_str = date_dt.strftime("%Y-%m-%d")

        # Call your "stats for a day" function
        day_stats = get_stats_for_day(date_str)
        if not day_stats:
            print(f"No stats found on {date_str}")
            continue
        total_legs = day_stats["total_legs"]
        print(f"Found stats for {total_legs} legs on {date_str}")
        # Store in a structured list
        results.append({
            "date": date_str,
            "stats": day_stats
        })

    return results
```

```python
import pandas as pd

# Suppose stats_list is the list shown above
def create_dataframe_from_stats(stats_list):
    # Flatten each record so that `date` and stats keys become columns
    records = []
    for entry in stats_list:
        row = {"date": entry["date"]}
        row.update(entry["stats"])  # merges all stats into the row
        records.append(row)

    # Create a DataFrame
    df = pd.DataFrame(records)
    
    # Convert "date" column to actual datetime objects
    df["date"] = pd.to_datetime(df["date"], format="%Y-%m-%d")
    
    # Set date as the index
    df.set_index("date", inplace=True)
    
    return df

df = create_dataframe_from_stats(recent_stats)
df.head()
```
